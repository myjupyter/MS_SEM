```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
a = -20
b = 30
XX = np.linspace(a,b,n)
I = 100
```

```python
XX
```

```python
ee = np.random.normal(0,1,n) 
AA = np.array([2,-2,1])
YY = np.polyval(AA, XX)
DATA = YY + I*ee
BB = np.polyfit(XX,DATA,len(AA)-1)
ZZ = np.polyval(BB, XX)
```

```python
plt.figure(dpi = 150)
plt.plot(XX, DATA, 'o')
plt.plot(XX, YY, 'y-',linewidth = 5)
plt.plot(XX, ZZ, 'k-' )
plt.figtext(0.92, 0.7, "N=%s\nAA=%s,%s,%s\nBB=%.3s, %.3s, %.3s \nI = %s"% (n,AA[0],AA[1],AA[2],
                                                                             BB[0],BB[1],BB[2],I))
plt.grid(True)
```
